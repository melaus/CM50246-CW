```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from generate_unsupervised_data import generate_unsupervised_data, plot_covar_ellipse
import gmm
import kmeans
```

# Generating 3 sets of datapoints
- pts_0: dimension = 2, clusters = 2,  points = 500
- pts_1: dimension = 2, clusters = 5,  points = 500
- pts_2: dimension = 2, clusters = 10, points = 5000

```python
pts_0, true_dist_0 = generate_unsupervised_data(num_dimensions=2,
                                                num_clusters=2,
                                                num_points=500,
                                                plot_data=True)
```

```python
pts_1, true_dist_1 = generate_unsupervised_data(num_dimensions=2,
                                                num_clusters=5,
                                                num_points=500,
                                                plot_data=True)
```

```python
pts_2, true_dist_2 = generate_unsupervised_data(num_dimensions=2,
                                                num_clusters=10,
                                                num_points=5000,
                                                plot_data=True)
```

# K-Means Plots

```python
# pts_0 K-means plot
title = 'K-means [d=2, k=2, pts=500]'
init_centroids_0 = kmeans.generate_centroids(pts_0.T, 2)
pts_min_dis_0, new_cen_list_0 = kmeans.kmeans(data=pts_0.T, centroids=init_centroids_0)
kmeans.plot_outcome(pts_min_dis_0, new_cen_list_0, title)
```

```python
# pts_1 K-means plot
title = 'K-means [d=2, k=5, pts=500]'
init_centroids_1 = kmeans.generate_centroids(pts_1.T, 5)
pts_min_dis_1, new_cen_list_1 = kmeans.kmeans(data=pts_1.T, centroids=init_centroids_1)
kmeans.plot_outcome(pts_min_dis_1, new_cen_list_1, title)
```

```python
# pts_1 K-means plot
title = 'K-means [d=2, k=10, pts=5000]'
init_centroids_2 = kmeans.generate_centroids(pts_2.T, 10)
pts_min_dis_2, new_cen_list_2 = kmeans.kmeans(data=pts_2.T, centroids=init_centroids_2)
kmeans.plot_outcome(pts_min_dis_2, new_cen_list_2, title)
```

# GMM plots

```python
# pts_0 in GMM [random init]
title = 'GMM [random] [d=2, k=2, pts=500]'
num_pts_0, num_k_0,  = 500, 2 
is_kmeans_0 = False
weights_0, means_0, covars_0 = gmm.initialise_parameters(num_pts_0, num_k_0, is_kmeans_0, pts_0.T)
gmm.gmm(2, pts_0.T, weights_0, means_0, covars_0, title)
plt.show()
```

```python
# pts_0 in GMM [K-means init]
title = 'GMM [k-means] [d=2, k=2, pts=500]'
num_pts_0, num_k_0,  = 500, 2
is_kmeans_0 = True
weights_0, means_0, covars_0 = gmm.initialise_parameters(num_pts_0, num_k_0, is_kmeans_0, pts_0.T)
gmm.gmm(2, pts_0.T, weights_0, means_0, covars_0, title)
plt.show()
```

```python
# pts_0 in GMM [K-means init] using an input c of 5
title = 'GMM [k-means] [d=2, k=2, pts=500] [using input k=5]'
num_pts_0, num_k_0,  = 500, 5 
is_kmeans_0 = True
weights_0, means_0, covars_0 = gmm.initialise_parameters(num_pts_0, num_k_0, is_kmeans_0, pts_0.T)
gmm.gmm(2, pts_0.T, weights_0, means_0, covars_0, title)
plt.show()
```
